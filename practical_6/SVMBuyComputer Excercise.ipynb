Support Vector Machine (SVM) Excercise

Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing the dataset

```python
dataset = pd.read_csv('Buy_Computer.csv')
```

```python
print(dataset)
```

Feature Scaling

```python
from sklearn.preprocessing import LabelEncoder
comp_data_num = dataset.apply(LabelEncoder().fit_transform)
data = np.array(comp_data_num.iloc[:, 1:5])
target = np.array(comp_data_num.iloc[:, 5])
```

```python
print(data)
```

Training the SVM model on the data set

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(data, target)
```

Creating new data to predict on

```python
test_data = {
    'id':[11],
    'age':[27],
    'income':['high'],
    'student':['no'],
    'credit_rating':['excellent']}

```

```python
#load test data into a DataFrame object:
test_df = pd.DataFrame(test_data)
```

```python
test_df = test_df.apply(LabelEncoder().fit_transform)
```

```python
testdata = np.array(test_df.iloc[:, 1:5])
```

```python
print(classifier.predict(testdata))
```
