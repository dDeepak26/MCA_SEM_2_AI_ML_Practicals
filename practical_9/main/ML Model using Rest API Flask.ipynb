```python

import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import requests, json
```

```python
#Load the data
data=pd.read_csv("titanic.csv")
data.describe()
```

```python
data.head()
```

```python
#dropping unnecessary features and encoding
le=LabelEncoder()
data['Sex']=le.fit_transform(data['Sex'].astype(str))
data['Embarked']=le.fit_transform(data['Embarked'].astype(str))
data.drop(['PassengerId'],axis=1,inplace=True)
data.drop(['Name'],axis=1,inplace=True)
data.drop(['Ticket'],axis=1,inplace=True)
data.drop(['Cabin'],axis=1,inplace=True)
data.drop(['Age'],axis=1,inplace=True)
data.head()
```

```python
#Divide the dataset into features and target variables
X=data.iloc[:,1:]
Y=data['Survived'].to_numpy()
print(X)
```

```python
#Fit the training data into the Random Forest Classifier
model=RandomForestClassifier().fit(X, Y)
```

Finally, we can serve the persisted model using a web framework. The following code creates a REST API using Flask. This file is hosted in a different environment, often in a cloud server.

```python
import joblib

joblib.dump(model, 'model.pkl')
```
