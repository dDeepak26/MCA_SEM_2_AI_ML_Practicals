```python
# Scientific and vector computation for python
import numpy as np
from sklearn import linear_model

import pandas as pd
# Plotting library
from matplotlib import pyplot

# Load data
data = pd.read_csv('KMeans_Dataset.csv')
print(data.head())

x = data['INCOME']
y= data['SPEND']

from sklearn.neighbors import KNeighborsClassifier

# Now we fit the KNN algorithm with K=1:

knn = KNeighborsClassifier(n_neighbors=1)

xarr = list(zip(x))

knn.fit (xarr, y)

# Predict the spend for income = 202

x_test = np.array([202])

prediction=knn.predict(x_test.reshape(-1, 1))

print(prediction)
```
